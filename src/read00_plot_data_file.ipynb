```python
datafolder = "long_1551_2/"
raw_data=[]
for file in os.listdir(datafolder)[:15]:
    data = ""
    with open(f"{datafolder}{file}","r") as f:
        data = f.readlines()
    raw_data.append( [int(d.split(",")[0]) for d in data[25:]] )
```

```python
plt.plot(np.array(raw_data).T);
#plt.yscale("log")
#plt.xscale("log")
#plt.xlim(3e3)
plt.show()
```
