```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(9,6))
plt.hist(np.random.normal(loc=8,scale=4**0.5,size=5000000),bins=100,density=True,histtype="step",lw=3,label="Sub-poissonian, variance<mean")
plt.hist(np.random.normal(loc=8,scale=8**0.5,size=5000000),bins=100,density=True,histtype="step",lw=3,label="Ideal poisson, variance=mean")
plt.hist(np.random.normal(loc=8,scale=15**0.5,size=5000000),bins=100,density=True,histtype="step",lw=3,label="Super-poissonian, variance>mean")
plt.legend()
plt.xlim(0,16)
plt.title("Simulated poissonian distributions")
plt.ylabel("Frequency")
plt.xlabel("Counts")
plt.savefig("poissonian_distribution.png")
```
